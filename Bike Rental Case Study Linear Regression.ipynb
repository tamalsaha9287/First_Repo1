```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing dataset
bike_rental = pd.read_csv('day.csv')
```

```python
#Check the top 5 rows
bike_rental.head()
```

```python
#data understanding
bike_rental.shape
```

```python
#Converting 'weathersit' into a string variable
bike_rental['weathersit'] = bike_rental['weathersit'].astype('str')
```

```python
bike_rental.describe()
```

```python
bike_rental.head()
```

```python
# Get the dummy variables for the feature 'weathersit' and store it in a new dataset - 'Weather_Condition'

Weather_Condition = pd.get_dummies(bike_rental['weathersit'])
```

```python
Weather_Condition.head()
```

```python
#Weather_Condition.weathersit1 = Weather_Condition['1']
Weather_Condition.rename(columns = {'1':'w1','2':'w2','3':'w3'}, inplace = True)
```

```python
Weather_Condition.head()
```

```python
# Drop the first column from status df using 'drop_first = True'
Weather_Condition = pd.get_dummies(bike_rental['weathersit'], drop_first = True)
Weather_Condition.head()
```

```python
# Interpretation of w2 & w3
# 00: weathersit = 1
# 10: weathersit = 2
# 01: weathersit = 3
Weather_Condition.rename(columns = {'2':'w2','3':'w3'}, inplace = True)
Weather_Condition.head()
```

```python
# Add the results to the original housing dataframe
bike_rental = pd.concat([bike_rental, Weather_Condition], axis = 1)
```

```python
bike_rental.head()
```

```python
# Drop 'weathersit' as we have created the dummies for it and drop the first column
bike_rental.drop(['weathersit'], axis = 1, inplace = True)

bike_rental.head()
```

```python
# Get the dummy variables for the feature 'season' and store it in a new dataset - 'Season_Ind'

Season_Ind = pd.get_dummies(bike_rental['season'], drop_first = True)
Season_Ind.head()
```

```python
# Season_Ind column value renaming
Season_Ind.rename(columns = {2:'s2',3:'s3',4:'s4'}, inplace = True)
```

```python
# 000: Season_Ind = s1
# 100: Season_Ind = s2
# 010: Season_Ind = s3
# 001: Season_Ind = s4
Season_Ind.head()
```

```python
# Add the results to the original bike rental dataframe
bike_rental = pd.concat([bike_rental, Season_Ind], axis = 1)
bike_rental.head()
```

```python
# Drop 'season' as we have created the dummies for it
bike_rental.drop(['season'], axis = 1, inplace = True)
bike_rental.head()
```

```python
# defining a formula to classify weekdays as Weekend_Flag binary variable
# 1: represents weekend

def weekend(x):
    if x in (4,5): return 1
    else: return 0
```

```python
bike_rental['Weekend_Flag'] = bike_rental.weekday.apply(weekend)
```

```python
bike_rental.head()
```

```python
# Drop few redundant variables like instant, casual, registered
bike_rental.pop('instant')
bike_rental.pop('casual')
bike_rental.pop('registered')
bike_rental.shape
```

```python
# Data split into training and test datasets
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
bike_rental_train, bike_rental_test = train_test_split(bike_rental, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
bike_rental_train.shape
```

```python
bike_rental_test.shape
```

```python
# Min Max Scalling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed']

bike_rental_train[num_vars] = scaler.fit_transform(bike_rental_train[num_vars])
bike_rental_train.head()
```

```python
bike_rental_train.describe()
```

```python
# Checking the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (16, 10))
sns.heatmap(bike_rental_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

## Drop highly correlated variables

## Running first model (lm1) using maximum variables

### Dividing into X and Y sets for the model building

```python
y_train = bike_rental_train.pop('cnt')
X_train = bike_rental_train
```

```python
y_train.head()
```

```python
X_train.head()
```

## RFE (Recursive feature elimination)

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=10)  
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

## Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#see the summary of our linear model
print(lm.summary())
```

## VIF calculation

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
X_train_new.columns
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Remove temp variable

```python
X_train_new1 = X_train_new.drop(['temp'], axis=1)
```

```python
X_train_new1
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm1 = sm.add_constant(X_train_new1)
```

```python
lm1 = sm.OLS(y_train,X_train_lm1).fit()   # Running the linear model
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_lm2
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Remove const due to high VIF
X_train_new3 = X_train_lm2.drop(['const'], axis=1)
```

```python
X_train_new3.head()
```

```python
X_train_new3.columns
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new3
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
